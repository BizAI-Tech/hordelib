```python
np.set_printoptions(threshold=np.inf, linewidth=80, suppress=True, precision=16)
```

```python
def print_state_dict(state_dict):
    print("{")
    for key, value in state_dict.items():
        value_np = value.numpy()
        value_str = np.array2string(value_np, separator=', ', suppress_small=False, precision=16)
        print(f'"{key}": {value_str}, \n')
    print("}")
```

```python
print_state_dict(model.model.state_dict())
```
